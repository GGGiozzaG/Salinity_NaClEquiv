# Equivalent NaCl Salinity Calculator
This notebook allows calculation of equivalent NaCl salinity based on ion concentration.
Calculation uses as reference values from Schlumberger Table (2013).

## Libraries

```python
import os
import numpy as np                                        # arrays and matrix math
import matplotlib.pyplot as plt                           # plotting
import pandas as pd
from scipy.interpolate import interp1d
```

```python
# Define global x-axis: First value is 0, followed by log scale from 10 to 300000
log_x = np.logspace(np.log10(10), np.log10(300000), num=500)  # Logarithmic scale from 10 to 300000
global_x = np.insert(log_x, 0, 0)  # Insert 0 at the beginning

# Folder containing the CSV files
folder_path = os.path.abspath(os.path.join(os.getcwd(), "Database"))

# Initialize dataframe with the global x-axis
df_result = pd.DataFrame({'x': global_x})

# Loop through all CSV files in the folder
for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path, file)
        
        # Read the CSV file (assuming no headers)
        data = pd.read_csv(file_path, header=None)
        
        x_data = data.iloc[:, 0].values  # First column as x-axis
        y_data = data.iloc[:, 1].values  # Second column as y-axis

        # Ensure data is sorted by x-values (if not already sorted)
        sorted_indices = np.argsort(x_data)
        x_data = x_data[sorted_indices]
        y_data = y_data[sorted_indices]

        # Define the interpolation function
        interp_func = interp1d(
            x_data, y_data, kind='linear', bounds_error=False, fill_value=np.nan
        )

        # Interpolate y-values onto the global x-axis
        y_interp = interp_func(global_x)

        # Handle values below the minimum x_data
        min_x = x_data[0]
        min_y = y_data[0]
        y_interp[global_x < min_x] = min_y  # Assign min_y for x-values below min_x

        # Store interpolated values in the result dataframe
        col_name = f"multi_{os.path.splitext(file)[0]}"  # Prefix "multi_" and remove file extension
        df_result[col_name] = y_interp

df_result.head()
```

```python
plt.plot(df_result['x'],df_result['multi_Ca'])
plt.plot(df_result['x'],df_result['multi_CO3'])
plt.plot(df_result['x'],df_result['multi_HCO3'])
plt.plot(df_result['x'],df_result['multi_K'])
plt.plot(df_result['x'],df_result['multi_Mg'])
plt.plot(df_result['x'],df_result['multi_SO4'])
plt.xscale('log'); plt.xlim(10,300000); plt.ylim(-0.5,2)
plt.xlabel('Concentration (ppm or mg/l)'); plt.ylabel('Multiplier (-)')
```

```python
def EqNaCl(df, m_Na, m_Cl, m_Ca = 0, m_Mg = 0, m_K = 0, m_SO4 = 0, m_CO3 = 0, m_HCO3 = 0, m_NH4 = 0, m_NO3 = 0, m_Br = 0, m_I = 0):
    interp = interp1d(
            df['x'], df['multi_Ca'], kind='linear', bounds_error=False, fill_value=np.nan
        )
    k_Ca = interp(m_Ca)
    interp = interp1d(
            df['x'], df['multi_Mg'], kind='linear', bounds_error=False, fill_value=np.nan
        )
    k_Mg = interp(m_Mg)
    interp = interp1d(
            df['x'], df['multi_K'], kind='linear', bounds_error=False, fill_value=np.nan
        )
    k_K = interp(m_K)
    interp = interp1d(
            df['x'], df['multi_SO4'], kind='linear', bounds_error=False, fill_value=np.nan
        )
    k_SO4 = interp(m_SO4)
    interp = interp1d(
            df['x'], df['multi_CO3'], kind='linear', bounds_error=False, fill_value=np.nan
        )
    k_CO3 = interp(m_CO3)
    interp = interp1d(
            df['x'], df['multi_HCO3'], kind='linear', bounds_error=False, fill_value=np.nan
        )
    k_HCO3 = interp(m_HCO3)
    NaCl = np.max([m_Na,m_Cl]) + m_Ca*k_Ca + m_Mg*k_Mg + m_K*k_K + m_SO4*k_SO4 + m_CO3*k_CO3 + m_HCO3*k_HCO3 + m_NH4*1.9 + m_NO3*0.55 + m_Br*0.44 + m_I*0.28
    return NaCl
print(EqNaCl(df_result, 585, 208, 76, 102, 69, 18, 1677, 0, 0, 0.61, 9.53, 0))
print(EqNaCl(df_result, 585, 208, 76, 102, 69, 18, 0, 1677, 0, 0.61, 9.53, 0))
print(EqNaCl(df_result, 585, 208, 76, 102, 69, 18, 0, 0, 0, 0.61, 9.53, 0))
print(EqNaCl(df_result, 10659, 19109, 409.25, 1289, 382.75, 1345.58, 0, 279.83, 0, 0, 35.52, 0))
```
